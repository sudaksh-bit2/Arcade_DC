```python
# Function to create custom segments for non-payers and whales to be upended later to segments from K means

def get_payers_non_whale(master_sub, plot=False):    
    Tot_spend = master_sub.groupby('acct_id').agg(tot_spend=('price','sum')).reset_index()
    Tot_spend['per_tot_spend_100']=round(Tot_spend.tot_spend.rank(pct=True, method='dense'),2)
    outlier_filter = Tot_spend.per_tot_spend_100 < 0.97 
    payer_filter = Tot_spend.tot_spend>0
    
    if plot==True:        
        k=0
        plt.figure()
        Tot_spend.groupby('per_tot_spend_100').tot_spend.mean().plot(logy=True, title='Total Spend for payers across 100 ranked buckets')
    
        plt.figure()
        Tot_spend[payer_filter & outlier_filter].groupby('per_tot_spend_100').tot_spend.mean().plot(logy=False, title='Total Spend for payers across 100 ranked buckets')
        
    payers_non_whale = Tot_spend.acct_id[payer_filter & outlier_filter]
    
    non_payer_dat = Tot_spend[~payer_filter][['acct_id']]
    non_payer_dat['segments']='non_payer'

    whale_dat = Tot_spend[~outlier_filter][['acct_id']]
    whale_dat['segments']='whale'

    custom_seg_data=non_payer_dat.append(whale_dat)
    
    return [payers_non_whale,custom_seg_data]

#Note: Remove multiple func calls for efficiency
# Create custom segments
payers_non_whale_0630 = get_payers_non_whale(master_sub_0630,plot=True)[0]
custom_seg_data_0630 = get_payers_non_whale(master_sub_0630)[1]

payers_non_whale_0731 = get_payers_non_whale(master_sub_0731)[0]
custom_seg_data_0731 = get_payers_non_whale(master_sub_0731)[1]

payers_non_whale_0831 = get_payers_non_whale(master_sub_0831)[0]
custom_seg_data_0831 = get_payers_non_whale(master_sub_0831)[1]


```

### Calculate where segments over or under index as compared to population by app category
- Calculate %age of transactions in overall population that belong to an app category(or any other dimension)
- Calculate %age of transactions in segment that belong to an app category(or any other dimension)
- Take the ratio to see if the segment over-indexes or under indexes in that category(or any cut of any dimension)

```python
def calculate_index(dimension, seg):
    print(dimension)
    # calculate spend across app_category by segment
    spend_cat=master_seg.pivot_table(index='segments',columns=dimension,values='price',aggfunc='sum').reset_index()

    # calculate total spend by app_category
    totals=master_seg.pivot_table(columns=dimension,values='price',aggfunc='sum').reset_index().rename(columns={'index':'segments'})

    # Append both together
    spend_cat=spend_cat.append(totals)

    # Calculate total spend across segments
    spend_cat['total_spend']= spend_cat.iloc[:,1:].sum(axis=1).values

    # Calculate percentage spend in each app_category by segment
    for i in spend_cat.columns[1:-1]:
        spend_cat[i+'_spend_perc']=spend_cat[i]/spend_cat.total_spend
        spend_cat.drop(i, axis=1,inplace=True)

    spend_cat = spend_cat.reset_index(drop=True)
    spend_cat_T=spend_cat.melt('segments').pivot(index=dimension,columns='segments',values='value')\
    .drop('total_spend',axis=0)
    (spend_cat_T[0]/spend_cat_T.price).plot.barh(title='index of spend participation by category in segment 0',
                                             log=False,
                                             xlim=0.6)
    plt.figure()
    (spend_cat_T[1]/spend_cat_T.price).plot.barh(title='index of spend participation by category in segment 1',
                                         log=False,
                                         xlim=0.8)
    plt.figure()
    (spend_cat_T[2]/spend_cat_T.price).plot.barh(title='index of spend participation by category in segment 2',
                                     log=False,
                                     xlim=0.6)
    plt.figure()
    (spend_cat_T['whale']/spend_cat_T.price).plot.barh(title='index of spend participation by category in segment whale',
                                             log=False,
                                             xlim=0)


calculate_index('content_type',2)
```
