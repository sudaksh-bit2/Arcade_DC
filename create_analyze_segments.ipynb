### Create custom hard segments based on spend behaviour

```python
def get_payers_non_whale(master_sub, plot=False):    
    Tot_spend = master_sub.groupby('acct_id').agg(tot_spend=('price','sum')).reset_index()
    Tot_spend['per_tot_spend_100']=round(Tot_spend.tot_spend.rank(pct=True, method='dense'),2)
    outlier_filter = Tot_spend.per_tot_spend_100 < 0.97 
    payer_filter = Tot_spend.tot_spend>0
    
    if plot==True:        
        k=0
        plt.figure()
        Tot_spend.groupby('per_tot_spend_100').tot_spend.mean().plot(logy=True, title='Total Spend for all accts across 100 ranked buckets')
    
        plt.figure()
        Tot_spend[payer_filter & outlier_filter].groupby('per_tot_spend_100').tot_spend.mean().plot(logy=False, title='Total Spend for all accts across 100 ranked buckets')
        
    payers_non_whale = Tot_spend.acct_id[payer_filter & outlier_filter]
    
    non_payer_dat = Tot_spend[~payer_filter][['acct_id']]
    non_payer_dat['segments']='non_payer'

    whale_dat = Tot_spend[~outlier_filter][['acct_id']]
    whale_dat['segments']='whale'

    custom_seg_data=non_payer_dat.append(whale_dat)
    
    return [payers_non_whale,custom_seg_data]

#Note: Remove multiple func calls
payers_non_whale_0630 = get_payers_non_whale(master_sub_0630,plot=True)[0]
custom_seg_data_0630 = get_payers_non_whale(master_sub_0630)[1]

payers_non_whale_0731 = get_payers_non_whale(master_sub_0731)[0]
custom_seg_data_0731 = get_payers_non_whale(master_sub_0731)[1]

payers_non_whale_0831 = get_payers_non_whale(master_sub_0831)[0]
custom_seg_data_0831 = get_payers_non_whale(master_sub_0831)[1]


```

#### What is the split of transactions between different App groups in each segment

```python
def seg_app_cat(app_cat):
    seg_games=master_seg[(master_seg.app_category==app_cat) & (master_seg.segments!='whale')].groupby('segments').agg(tot_game_spend=('price','sum'),
                                                                      users=('acct_id','nunique')).reset_index()
    seg_games['spend_per_user']=seg_games.tot_game_spend/seg_games.users

    plt.figure()
    plt.bar(x=seg_games.index, height=seg_games.spend_per_user,log=True, tick_label=seg_games.segments)
    plt.title('Spend_per_user- '+app_cat)
    plt.show()

for app_cat in master_seg.app_category.unique():
    seg_app_cat(app_cat)
```

#### What is the split of transactions between consumables/Subscription in each segment

```python
def seg_cont_type(cont_type):
    seg_cont=master_seg[(master_seg.content_type==cont_type) & (master_seg.segments!='whale')].groupby('segments').agg(tot_spend=('price','sum'),
                                                                      users=('acct_id','nunique')).reset_index()
    seg_cont['spend_per_user']=seg_cont.tot_spend/seg_cont.users

    plt.figure()
    plt.bar(x=seg_cont.index, height=seg_cont.spend_per_user,log=True, tick_label=seg_cont.segments)
    plt.title('Spend_per_user- '+cont_type)
    plt.show()

for content_type in master_seg.content_type.unique():
    seg_cont_type(content_type)
```
